## Problem Statement

This dataset contains the sign and symptpom data of newly diabetic or would be diabetic patient.

## Data Description

Data Set Information:

This dataset contains reports of diabetes-related symptoms of 520 persons. It includes data about peoples including symptoms that may cause diabetes. This dataset has been created from a direct questionnaire to people who have recently become diabetic, or who are still nondiabetic but having few or more symptoms. The data has been collected from the patients using direct questionnaire from Sylhet Diabetes Hospital of Sylhet, Bangladesh.

Attribute Information:

| Variables          | Description       |

### Outlier Treatment

```python
diabetes.Age.quantile(0.99)
```

```python
def clean_age(age):     
    if age>=0 and age<=75:  
        return age
    else:
        return np.nan
```

```python
diabetes['age_clean'] = diabetes['Age'].apply(clean_age)

print("'Age'")                                         
print("Minimum value: ", diabetes["age_clean"].min())
print("Maximum value: ",diabetes["age_clean"].max())
print("How many values are NaN?: ", pd.isnull(diabetes['age_clean']).sum())
print("Frequency table for age_clean:")
```

```python
diabetes["age_clean"] = diabetes["age_clean"].fillna(diabetes["age_clean"].mean())
```

```python
print("How many values are NaN?: ", pd.isnull(diabetes['age_clean']).sum())
```

```python
diabetes["age_clean"] = diabetes["age_clean"].astype(int)
```

```python
diabetes.drop(['Age'],axis=1,inplace=True)
```

```python
diabetes.rename(columns={'age_clean':'Age'},inplace=True)
```

## Machine Learning Modeling

### Converting catogorical value to number

```python
diabetes.columns
```

```python
cat_columns = ['Alopecia', 'Gender', 'Genital thrush', 'Irritability','Itching', 'Obesity', 'Polydipsia', 'Polyphagia',
               'Polyuria', 'delayed healing', 'muscle stiffness', 'partial paresis','sudden weight loss', 'visual blurring',
               'weakness','class']
diabetes = pd.get_dummies(diabetes,columns=cat_columns,drop_first=True) 
```

```python
diabetes.head()
```

### Splitting X and y into training and test datasets

```python
X = diabetes.iloc[:,:-1]
y = diabetes['class_Positive']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

### Scaling

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)

X_test_s = scaler.transform(X_test)
```

### Logistic Regression

```python
model_LOGR = LogisticRegression()
model_LOGR.fit(X_train_s,y_train)
```

```python
y_pred_train_LOGR = model_LOGR.predict(X_train_s)  

y_pred_test_LOGR = model_LOGR.predict(X_test_s) 
```

#### Model evaluation for Logistic Regression

```python
train_accuracy_LOGR = model_LOGR.score(X_train_s,y_train)
test_accuracy_LOGR = model_LOGR.score(X_test_s,y_test)
p_score_LOGR = metrics.precision_score(y_test,y_pred_test_LOGR)
r_score_LOGR = metrics.recall_score(y_test,y_pred_test_LOGR)
f1_score_LOGR = metrics.f1_score(y_test,y_pred_test_LOGR)
fp, tp, th = metrics.roc_curve(y_test,y_pred_test_LOGR)
auc_LOGR = metrics.auc(fp, tp)
```

```python
print("Train Accuracy: {}".format(round(train_accuracy_LOGR,3)))
print("Test Accuracy: {}".format(round(test_accuracy_LOGR,3)))
print("Precision Score: {}".format(round(p_score_LOGR,3)))
print("Recall Score: {}".format(round(r_score_LOGR,3)))
print("F1 Score: {}".format(round(f1_score_LOGR,3)))
print("AUC: {}".format(round(auc_LOGR,3)))
print("==============Classification Report=============================")
print(metrics.classification_report(y_test,y_pred_test_LOGR))
print("==============Confusion Matrix=============================")
print(metrics.confusion_matrix(y_test,y_pred_test_LOGR))
```

#### AUC Curve for Logistic Regression Algorithm

```python
probs = model_LOGR.predict_proba(X_test_s)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc_LOGR = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc_LOGR)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### Decision Tree Classifier

```python
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train_s, y_train)
```

```python
y_pred_train_DTC = model_DTC.predict(X_train)  

y_pred_test_DTC = model_DTC.predict(X_test) 
```

#### Model evaluation for Decision Tree Classifier

```python
train_accuracy_DTC = model_DTC.score(X_train_s,y_train)
test_accuracy_DTC = model_DTC.score(X_test_s,y_test)
p_score_DTC = metrics.precision_score(y_test,y_pred_test_DTC)
r_score_DTC = metrics.recall_score(y_test,y_pred_test_DTC)
f1_score_DTC = metrics.f1_score(y_test,y_pred_test_DTC)
fp, tp, th = metrics.roc_curve(y_test,y_pred_test_DTC)
auc_DTC = metrics.auc(fp, tp)
```

```python
print("Train Accuracy: {}".format(round(train_accuracy_DTC,3)))
print("Test Accuracy: {}".format(round(test_accuracy_DTC,3)))
print("Precision Score: {}".format(round(p_score_DTC,3)))
print("Recall Score: {}".format(round(r_score_DTC,3)))
print("F1 Score: {}".format(round(f1_score_DTC,3)))
print("AUC: {}".format(round(auc_DTC,3)))
print("==============Classification Report=============================")
print(metrics.classification_report(y_test,y_pred_test_DTC))
print("==============Confusion Matrix=============================")
print(metrics.confusion_matrix(y_test,y_pred_test_DTC))
```

#### AUC Curve for Decision Tree Algorithm

```python
probs = model_DTC.predict_proba(X_test_s)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc_DTC = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc_DTC)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
